```python
#This block will take all the json_dicts generated by Dallin and
#put them into an objects named allRecursiveDicts.
#This object looks like {'1915-1935': {'HarmVirtue': ['martyrs',...
#                                     {'HarmVice'}: ['...

directory = "json_dicts"
# Create an empty dictionary to store the Python dictionaries
allRecursiveDicts = {}

# Loop through each file in the directory
for filename in os.listdir(directory):
    # Construct the file path
    filepath = os.path.join(directory, filename)
    # Check if the file is a regular file
    if os.path.isfile(filepath):
        # Read in the contents of the file
        with open(filepath, 'r') as file:
            # Load the JSON data into a Python dictionary
            dictionaryTerms = json.load(file)
            # Extract the year range from the filename
            year_range = filename.split("_")[0]
            # Add the data to the dictionary with the year range as the key
            allRecursiveDicts[year_range] = dictionaryTerms
            
            
#This sorts the years in the dictionary to be in the right order.
chronological_order = sorted(allRecursiveDicts.keys())
placeholder = allRecursiveDicts
allRecursiveDicts = OrderedDict()
for yearRange in chronological_order:
    allRecursiveDicts[yearRange] = placeholder[yearRange]
        
#Lets take a look at the year ranges and make sure we have all of them, and that they are in the right order
for yearRange in allRecursiveDicts:
    print(yearRange)
```

```python
## This code tokenizes every document and puts it in the appropriate date range and saves it to disk using pickle.

#First, initialize the object to store all of the tokens across all the date ranges
counter_allPresTokens_allRanges = OrderedDict({yearRange: OrderedDict() for yearRange in allRecursiveDicts})

for yearRange in allRecursiveDicts:
    presidentsInYearRange = []
    yearLower = int(yearRange[:4])
    yearHigher = int(yearRange[-4:])
    for president in presidents:
        termStart, termEnd = min(presidents[president]), max(presidents[president])
        if yearLower <= termEnd and yearHigher >= termStart:
            presidentsInYearRange.append(president)
    sorted_presidents = sorted(presidentsInYearRange, key=lambda x: min(presidents[x]))
    counter_allPresTokens_allRanges[yearRange] = OrderedDict((president, Counter()) for president in sorted_presidents)

directory = 'parsedDocumentsPresidencyOnly'

#Add the tokens to the object
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isfile(f):
        J = open(directory+'/'+filename)
        presidentialSpeechFile = json.loads(J.read())
        J.close()
        # Get the year from the speech file
        year = presidentialSpeechFile['year']
        speaker = presidentialSpeechFile['speaker'].replace(' ', '')
        for yearRange in counter_allPresTokens_allRanges:
            #Pull the lower and higher years from the range
            yearLower = int(yearRange[:4])
            yearHigher = int(yearRange[-4:])
            #Put it in right year range
            if yearLower <= year <= yearHigher:
                for token in word_tokenize(presidentialSpeechFile['body']):
                    token = token.lower()
                    counter_allPresTokens_allRanges[yearRange][speaker][token] += 1

#Create the directory if it doesn't exist

if not os.path.exists("pickleFiles"):
    os.mkdir("pickleFiles")

#Save it to disk with pickle.

with open("pickleFiles/counter_allPresTokens_allRanges_dynamic.pickle", "wb") as handle:
    pickle.dump(counter_allPresTokens_allRanges, handle, protocol = pickle.HIGHEST_PROTOCOL)
    
Audio(sound_file, autoplay=True)
```

```python
#Takes about a minute
#This code creates and formats the dataframes that contain all the tokens by the president in the given year range.
#The last part in each block sums each of the tokens and formats it so that it only contains the president as the columns
#And the sum as the cells in the row.

df_allPresTokens_allRanges = OrderedDict({yearRange:pd.DataFrame(counter_allPresTokens_allRanges[yearRange]) 
                              for yearRange in counter_allPresTokens_allRanges})
df_sumOfAllToken_allRanges = OrderedDict({yearRange:pd.DataFrame(counter_allPresTokens_allRanges[yearRange]) 
                              for yearRange in counter_allPresTokens_allRanges})

for yearRange in df_allPresTokens_allRanges:
    df_allPresTokens_allRanges[yearRange].reset_index(inplace = True)
    df_allPresTokens_allRanges[yearRange].rename(columns={'index': 'token'}, inplace = True)
    df_sumOfAllToken_allRanges[yearRange] = df_allPresTokens_allRanges[yearRange].sum(axis = 0).drop('token')

#This part creates the dataframe with every president in the same dataframe 

df_sumOfAllTokens_allPres = pd.concat([df_sumOfAllToken_allRanges[yearRange] for yearRange in df_sumOfAllToken_allRanges], axis=1)
df_sumOfAllTokens_allPres = df_sumOfAllTokens_allPres.groupby(df_sumOfAllTokens_allPres.columns, axis=1).sum()
df_sumOfAllTokens_allPres = df_sumOfAllTokens_allPres.transpose().sum(axis = 0)
```

```python
#This does the same as the next cell, but only for exact word matches and not using regular expressions.

allMatchedIndices = {}

for yearRange in df_allPresTokens_allRanges:
    if yearRange not in allMatchedIndices:
        allMatchedIndices[yearRange] = {}
    for moralCategory in allRecursiveDicts[yearRange]:
        if moralCategory not in allMatchedIndices[yearRange]:
            allMatchedIndices[yearRange][moralCategory] = []
        for index, row in df_allPresTokens_allRanges[yearRange].iterrows():
            if row['token'] in allRecursiveDicts[yearRange][moralCategory]:
                allMatchedIndices[yearRange][moralCategory].append(index)
```

```python
## Initialize the indices and match all the tokens with the regular expression in the dataframes across all year ranges.

allMatchedIndices = {}

for yearRange in df_allPresTokens_allRanges:
    if yearRange not in allMatchedIndices:
        allMatchedIndices[yearRange] = {}
    for moralCategory in allRecursiveDicts[yearRange]:
        #Pull the terms from the dictionary for each moral category and compile an re to match the terms in the df
        terms_to_search_for = {term.lower() if '*' in term else re.escape(term.lower()) for term in allRecursiveDicts[yearRange][moralCategory]}
        regex = '|'.join(terms_to_search_for)
        pattern = r'\b'+r"\b|\b".join(terms_to_search_for)+r'\b'
        pattern = pattern.replace('*', '.*')
        compiled_pattern = re.compile(pattern)
        #Match the terms in the df
        match = df_allPresTokens_allRanges[yearRange]['token'].str.contains(compiled_pattern)
#         match = df_allPresTokens_allRanges[yearRange]['token'].str.match(regex)
        #Add the matched indices into the appropriate range of years and moral category of allMatchedIndices
        allMatchedIndices[yearRange][moralCategory] = df_allPresTokens_allRanges[yearRange][match].index.tolist()
        
# pprint(allMatchedIndices, compact = True)
```

```python
#This code creates dataframes for each of the ranges of years that contains the moral category along the rows and
#The presidents along the columns. The cells contain the sums for the respective president/category.
#It looks like this:
#   category	    GeorgeW.Bush	BarackObama	WilliamJ.Clinton
# 0	HarmVirtue	    14619.0	        10652.0	    3745.0
# 1	AuthorityVirtue	9240.0	        9326.0	    2495.0

#Init dataframe to hold the category, president, and sums
df_withAllCategories_2001_2016 = pd.DataFrame(columns=['president'] + ['sum'])
df_withAllCategories_1893_1912 = pd.DataFrame(columns=['president'] + ['sum'])
df_withAllCategories_1873_1892 = pd.DataFrame(columns=['president'] + ['sum'])
df_withAllCategories_1953_1972 = pd.DataFrame(columns=['president'] + ['sum'])
df_withAllCategories_1973_1992 = pd.DataFrame(columns=['president'] + ['sum'])
df_withAllCategories_1993_2000 = pd.DataFrame(columns=['president'] + ['sum'])
df_withAllCategories_1933_1952 = pd.DataFrame(columns=['president'] + ['sum'])
df_withAllCategories_1913_1932 = pd.DataFrame(columns=['president'] + ['sum'])

for year_range in allMatchedIndices:
    for moralCategory in allMatchedIndices[year_range]:
        #Iterate through all parts of the Matched_Indices object
        matched_indices = allMatchedIndices[year_range][moralCategory]
        #Find the appropriate range of years
        if year_range == '2001-2016':
            #Add teh sum of the columns onto the bottom of the df_withSum
            df_withSum = allPresTokens_df_2001_2016.loc[matched_indices, :].sum(axis = 0)
            #Formate it to go with the final dataframe 
            df_withSum = df_withSum.to_frame().transpose()
            df_withSum.insert(0, 'category', moralCategory, True)
            #Add df_withSum to the df_withAllCategories_ with teh appropriate range of years
            df_withAllCategories_2001_2016 = pd.concat([df_withAllCategories_2001_2016, df_withSum], ignore_index=True)
        elif year_range == '1893-1912':
            df_withSum = allPresTokens_df_1893_1912.loc[matched_indices, :].sum(axis = 0)
            df_withSum = df_withSum.to_frame().transpose()
            df_withSum.insert(0, 'category', moralCategory, True)
            df_withAllCategories_1893_1912 = pd.concat([df_withAllCategories_1893_1912, df_withSum], ignore_index=True)
        elif year_range == '1873-1892':
            df_withSum = allPresTokens_df_1873_1892.loc[matched_indices, :].sum(axis = 0)
            df_withSum = df_withSum.to_frame().transpose()
            df_withSum.insert(0, 'category', moralCategory, True)
            df_withAllCategories_1873_1892 = pd.concat([df_withAllCategories_1873_1892, df_withSum], ignore_index=True)
        elif year_range == '1953-1972':
            df_withSum = allPresTokens_df_1953_1972.loc[matched_indices, :].sum(axis = 0)
            df_withSum = df_withSum.to_frame().transpose()
            df_withSum.insert(0, 'category', moralCategory, True)
            df_withAllCategories_1953_1972 = pd.concat([df_withAllCategories_1953_1972, df_withSum], ignore_index=True)
        elif year_range == '1973-1992':
            df_withSum = allPresTokens_df_1973_1992.loc[matched_indices, :].sum(axis = 0)
            df_withSum = df_withSum.to_frame().transpose()
            df_withSum.insert(0, 'category', moralCategory, True)
            df_withAllCategories_1973_1992 = pd.concat([df_withAllCategories_1973_1992, df_withSum], ignore_index=True)
        elif year_range == '1993-2000':
            df_withSum = allPresTokens_df_1993_2000.loc[matched_indices, :].sum(axis = 0)
            df_withSum = df_withSum.to_frame().transpose()
            df_withSum.insert(0, 'category', moralCategory, True)
            df_withAllCategories_1993_2000 = pd.concat([df_withAllCategories_1993_2000, df_withSum], ignore_index=True)
        elif year_range == '1933-1952':
            df_withSum = allPresTokens_df_1933_1952.loc[matched_indices, :].sum(axis = 0)
            df_withSum = df_withSum.to_frame().transpose()
            df_withSum.insert(0, 'category', moralCategory, True)
            df_withAllCategories_1933_1952 = pd.concat([df_withAllCategories_1933_1952, df_withSum], ignore_index=True)
        elif year_range == '1913-1932':
            df_withSum = allPresTokens_df_1913_1932.loc[matched_indices, :].sum(axis = 0)
            df_withSum = df_withSum.to_frame().transpose()
            df_withSum.insert(0, 'category', moralCategory, True)
            df_withAllCategories_1913_1932 = pd.concat([df_withAllCategories_1913_1932, df_withSum], ignore_index=True)

#Finally, get rid of the extra columns in the final databases.
df_withAllCategories_2001_2016 = df_withAllCategories_2001_2016.drop(['president', 'sum', 'token'], axis = 1) 
df_withAllCategories_1893_1912 = df_withAllCategories_1893_1912.drop(['president', 'sum', 'token'], axis = 1) 
df_withAllCategories_1873_1892 = df_withAllCategories_1873_1892.drop(['president', 'sum', 'token'], axis = 1) 
df_withAllCategories_1953_1972 = df_withAllCategories_1953_1972.drop(['president', 'sum', 'token'], axis = 1) 
df_withAllCategories_1973_1992 = df_withAllCategories_1973_1992.drop(['president', 'sum', 'token'], axis = 1) 
df_withAllCategories_1993_2000 = df_withAllCategories_1993_2000.drop(['president', 'sum', 'token'], axis = 1) 
df_withAllCategories_1933_1952 = df_withAllCategories_1933_1952.drop(['president', 'sum', 'token'], axis = 1)
df_withAllCategories_1913_1932 = df_withAllCategories_1913_1932.drop(['president', 'sum', 'token'], axis = 1)
```

```python
#This code creates a dataframe (df_aggregrateAcrossAllCategories) that has each president in the columns, and each of the
#categories in the rows.It gets rid of all the date ranges, so it is just the sum for each president across all date ranges.
#It looks like:
#   category	    GroverCleveland	ChesterA.Arthur
#0	AuthorityVice	257.0	        88.0	
#1	AuthorityVirtue	378.0	        111.0

df_withAllCategories_noYearRanges = pd.concat([df_withAllCategories_2001_2016, df_withAllCategories_1893_1912, 
                                               df_withAllCategories_1873_1892, df_withAllCategories_1953_1972, 
                                               df_withAllCategories_1973_1992, df_withAllCategories_1993_2000, 
                                               df_withAllCategories_1933_1952, df_withAllCategories_1913_1932], axis=0)
df_withAllCategories_noYearRanges = df_withAllCategories_noYearRanges.set_index('category')
df_withAllCategories_noYearRanges.reset_index(inplace=True)
df_aggregrateAcrossAllCategories = df_withAllCategories_noYearRanges.groupby(['category']).sum()
df_aggregrateAcrossAllCategories.reset_index(inplace=True)

#This code creates a dataframe named df_normalizedSumsAllCategories that has the presidents in the columns and the 
#Moral categories in the rows. It then divides the sums of the moral categories by the total tokens used by that president.
#It looks like this:
#   category	    GroverCleveland	ChesterA.Arthur
#0	AuthorityVice	0.000523	    0.000747	
#1	AuthorityVirtue	0.000769	    0.000942

df_aggregrateAcrossAllCategories_withTokenSum = df_aggregrateAcrossAllCategories.append(df_sumOfAllTokens_allYearRanges, 
                                                                                        ignore_index = True)
df_aggregrateAcrossAllCategories_withTokenSum.iloc[-1,0] = 'Token_Sums'

df_normalizedSumsAllCategories = df_aggregrateAcrossAllCategories_withTokenSum
for row in range(df_normalizedSumsAllCategories.shape[0]):
  df_normalizedSumsAllCategories.iloc[row,1:] = df_normalizedSumsAllCategories.iloc[row,1:]/df_normalizedSumsAllCategories.iloc[-1][1:]

df_normalizedSumsAllCategories = df_normalizedSumsAllCategories.iloc[:-1,:]
```
